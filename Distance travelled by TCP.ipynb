```python
from robolink import *  # RoboDK API
from robodk import *    # Robot toolbox
import time
from math import sqrt

# Connect to RoboDK
RDK = Robolink()

# Select the robot
robot = RDK.ItemUserPick('Select a Robot', ITEM_TYPE_ROBOT)
if not robot.Valid():
    raise Exception("No robot selected or available")

# Set TCP to robot flange (no tool)
robot.setTool(Pose(0, 0, 0, 0, 0, 0))  # Reset to robot flange (no tool offset)

# Confirm the current TCP position in robot's local frame
current_tcp_local = robot.PoseTool()
print(f"TCP (local to robot flange): {current_tcp_local}")

# Convert the local TCP position to the world frame (robot base coordinate system)
tcp_world = robot.Pose() * current_tcp_local  # Transformation to world frame
print(f"TCP Position (relative to world origin): {tcp_world.Pos()}")

# Select a program to run
program = RDK.ItemUserPick('Select a Program', ITEM_TYPE_PROGRAM)
if not program.Valid():
    raise Exception("No valid program selected")

print(f"Program selected: {program.Name()}")

# Run the selected program
program.RunProgram()

# Initialize variables for distance calculation
distance = 0
prev_pose = None  # Initialize previous position variable

# Monitor program execution and print TCP position while the program is running
while program.Busy():
    RDK.Update()  # Refresh simulation
    pose_local = robot.PoseTool()  # Get the current TCP pose in robot's local frame
    
    # Convert to world frame
    pose_world = robot.Pose() * pose_local  # Transform the pose to world frame
    current_pos = pose_world.Pos()  # Get the current position in world frame
    print(f"Current TCP Position (World frame): {current_pos}")

    # Calculate the distance if there was a previous pose
    if prev_pose is not None:
        delta = sqrt(
            (current_pos[0] - prev_pose[0]) ** 2 +
            (current_pos[1] - prev_pose[1]) ** 2 +
            (current_pos[2] - prev_pose[2]) ** 2
        )
        distance += delta

    prev_pose = current_pos  # Update the previous pose

    # Optional: Add a short delay to control the printing speed
    time.sleep(0.1)

print("Program finished.")
# Output the total distance
print(f"Total TCP Distance Travelled: {distance:.2f} mm")
```
